```python
import re

def fix_hebrew_pdf_line(line):
    reversed_line = line[::-1]
    
    # Fix 4-digit years/numbers that got reversed
    reversed_line = re.sub(
        r"\b(\d{4})\b",
        lambda m: m.group(1)[::-1],
        reversed_line
    )
    
    return reversed_line

fixed_text = "\n".join(
    fix_hebrew_pdf_line(line) 
    for line in text.split("\n")
)

print(fixed_text[:1000])
```

```python
# ===============================
# שלב 4: חילוץ טקסט מלא מהדוח ותיקון עברית
# הסבר: קורא את כל עמודי ה-PDF, מתקן כיוון עברית ומספרים,
# ושומר כל עמוד יחד עם מספר עמוד ומקור (לשימוש ב-RAG)
# ===============================

import pdfplumber
import re

file_path = "C:/Users/zoars/young_hub/data/raw_reports/Young_Statistical_2025.pdf"

def fix_hebrew_pdf_line(line):
    reversed_line = line[::-1]
    
    # תיקון מספרים (כמו שנים שהתבלבלו)
    reversed_line = re.sub(
        r"\b(\d{4})\b",
        lambda m: m.group(1)[::-1],
        reversed_line
    )
    
    return reversed_line

documents = []

with pdfplumber.open(file_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        
        if text:
            fixed_text = "\n".join(
                fix_hebrew_pdf_line(line) 
                for line in text.split("\n")
            )
            
            documents.append({
                "text": fixed_text,
                "page": i + 1,
                "source": "Young_Statistical_2025"
            })

# בדיקה ראשונית
print(len(documents))
print(documents[0]["text"][:500])
```

```python
# ===============================
# שלב 5: בדיקת איכות חילוץ בעמוד אמצעי
# הסבר: מציג טקסט מעמוד פנימי כדי לוודא שהדוח קריא גם באמצע,
# לפני שמחלקים את הטקסט לקטעים קטנים עבור RAG
# ===============================

page_number = 150

page_doc = next(
    doc for doc in documents 
    if doc["page"] == page_number
)

print(page_doc["text"][:1500])
```

```python
# ===============================
# שלב 6: תיקון שגיאות חילוץ נפוצות
# הסבר: מתקן טעויות שחוזרות בטקסט המחולץ, למשל OECD שהתהפך ל-DCEO
# ===============================

def clean_extracted_text(text):
    replacements = {
        "DCEO": "OECD",
        "צהיריל בי ראל": "צעירים בישראל",
        "̶–": "–"
    }
    
    for wrong, correct in replacements.items():
        text = text.replace(wrong, correct)
    
    return text

for doc in documents:
    doc["text"] = clean_extracted_text(doc["text"])

# בדיקה חוזרת של אותו עמוד
page_doc = next(
    doc for doc in documents 
    if doc["page"] == 150
)

print(page_doc["text"][:1500])
```

```python
# ===============================
# שלב 7: חלוקת הדוח לקטעי טקסט
# הסבר: מחלק כל עמוד לקטעים קטנים עם חפיפה קלה,
# כדי לשפר את יכולת החיפוש והשליפה של מערכת ה-RAG
# ===============================

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=150,
    separators=["\n\n", "\n", ".", " ", ""]
)

chunks = []

for doc in documents:
    page_chunks = text_splitter.split_text(doc["text"])
    
    for chunk_id, chunk_text in enumerate(page_chunks):
        chunks.append({
            "text": chunk_text,
            "page": doc["page"],
            "source": doc["source"],
            "chunk_id": chunk_id
        })

print(f"Total chunks: {len(chunks)}")
print(chunks[0])
```

```python
# ===============================
# בדיקת אורך chunk
# הסבר: בודק כמה תווים יש בפועל בקטע
# ===============================

print(len(chunks[0]["text"]))
print(len(chunks[1]["text"]))
```

```python
# ===============================
# בדיקת אורכי chunks בכל הדוח
# הסבר: מציג את האורך המינימלי, הממוצע והמקסימלי של הקטעים
# ===============================

chunk_lengths = [len(chunk["text"]) for chunk in chunks]

print("Number of chunks:", len(chunk_lengths))
print("Shortest chunk:", min(chunk_lengths))
print("Average chunk length:", round(sum(chunk_lengths) / len(chunk_lengths), 1))
print("Longest chunk:", max(chunk_lengths))
```

```python
# ===============================
# שלב 8: סינון chunks קצרים מדי
# הסבר: מסיר קטעים קצרים שלא תורמים להבנת התוכן
# ===============================

MIN_CHUNK_LENGTH = 100  # אפשר לכוון בעתיד

filtered_chunks = [
    chunk for chunk in chunks 
    if len(chunk["text"].strip()) >= MIN_CHUNK_LENGTH
]

print("Original chunks:", len(chunks))
print("Filtered chunks:", len(filtered_chunks))
```

```python
# ===============================
# שלב 9: יצירת embeddings לכל chunk
# הסבר: ממיר כל קטע טקסט לוקטור מספרי באמצעות מודל שפה
# ===============================

from langchain_community.embeddings import HuggingFaceEmbeddings

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2"
)

# יצירת רשימת טקסטים בלבד
texts = [chunk["text"] for chunk in filtered_chunks]

# יצירת embeddings
embeddings = embedding_model.embed_documents(texts)

print(len(embeddings))
print(len(embeddings[0]))
```

```python
# ===============================
# שלב 12: תיקון מספרים, טווחי גיל ואחוזים שהתהפכו
# הסבר: משפר את איכות הטקסט לאחר חילוץ PDF בעברית,
# במיוחד במספרים כמו גילאים ואחוזים
# ===============================

def fix_reversed_numbers(text):
    replacements = {
        "81-42": "18–24",
        "52-92": "25–29",
        "%6.31": "13.6%",
        "%1.61": "16.1%",
        "%7.4": "4.7%",
        "%1.13": "31.1%",
        "%2.8": "8.2%",
        "%3.33": "33.3%",
        "%8.61": "16.8%",
        "%2.51": "15.2%",
        "51": "15",
        "41": "14"
    }
    
    for wrong, correct in replacements.items():
        text = text.replace(wrong, correct)
    
    return text

for doc in documents:
    doc["text"] = fix_reversed_numbers(doc["text"])

for chunk in filtered_chunks:
    chunk["text"] = fix_reversed_numbers(chunk["text"])
```

```python
# ===============================
# שלב 16 (ללא מודל): הצגת תשובה מבוססת מקורות
# הסבר: מחזיר את הקטעים הכי רלוונטיים בצורה מסודרת,
# בלי להשתמש ב-LLM חיצוני
# ===============================

query = "צעירים שאינם עובדים ואינם לומדים בישראל"

results = vector_db.similarity_search(query, k=3)

for i, result in enumerate(results, start=1):
    print(f"\n### מקור {i}")
    print(f"דוח: {result.metadata['source']}")
    print(f"עמוד: {result.metadata['page']}")
    print(result.page_content[:800])
```

```python
# ===============================
# שלב 16: פונקציית חיפוש בסיסית במרכז הידע
# הסבר: יוצרת פונקציה שמקבלת שאלה בעברית,
# מחפשת קטעים רלוונטיים ומחזירה מקור, עמוד וטקסט
# ===============================

def search_knowledge_base(query, k=5):
    results = vector_db.similarity_search(query, k=k)
    
    for i, result in enumerate(results, start=1):
        print(f"\n--- תוצאה {i} ---")
        print(f"דוח: {result.metadata['source']}")
        print(f"עמוד: {result.metadata['page']}")
        print(result.page_content[:1000])
```

```python
# ===============================
# שלב 17: בדיקת שאלות לדוגמה
# הסבר: בודק האם החיפוש מחזיר קטעים רלוונטיים
# עבור שאלות אמיתיות שמשתמשים עשויים לשאול
# ===============================

search_knowledge_base(
    "מה ידוע על צעירים שאינם עובדים ואינם לומדים?",
    k=5
)
```

```python
# ===============================
# שלב 18: חיפוש משולב ומדויק יותר
# הסבר: מחפש סמנטית, אבל נותן עדיפות לתוצאות שמכילות
# את הביטוי המדויק "אינם עובדים ואינם לומדים"
# ===============================

def search_knowledge_base_precise(query, exact_phrase=None, k=5, fetch_k=20):
    raw_results = vector_db.similarity_search(query, k=fetch_k)
    
    if exact_phrase:
        exact_results = [
            result for result in raw_results
            if exact_phrase in result.page_content
        ]
        
        other_results = [
            result for result in raw_results
            if exact_phrase not in result.page_content
        ]
        
        results = exact_results + other_results
    else:
        results = raw_results
    
    for i, result in enumerate(results[:k], start=1):
        print(f"\n--- תוצאה {i} ---")
        print(f"דוח: {result.metadata['source']}")
        print(f"עמוד: {result.metadata['page']}")
        print(result.page_content[:1000])
```

```python
# ===============================
# שלב 19: בדיקת חיפוש עם ביטוי מדויק
# הסבר: בודק האם התוצאות הרלוונטיות ביותר עולות לראש הרשימה
# ===============================

search_knowledge_base_precise(
    query="מה ידוע על צעירים שאינם עובדים ואינם לומדים?",
    exact_phrase="אינם עובדים ואינם לומדים",
    k=5,
    fetch_k=20
)
```

```python
# ===============================
# שלב 20: דירוג תוצאות לפי סימנים חזקים
# הסבר: נותן ניקוד גבוה יותר לקטעים שיש בהם גם את הביטוי המרכזי
# וגם סימנים של נתונים, כמו שנים, אחוזים, OECD וישראל
# ===============================

def rank_result(result, exact_phrase):
    text = result.page_content
    score = 0
    
    if exact_phrase in text:
        score += 5
    if "OECD" in text:
        score += 3
    if "ישראל" in text:
        score += 2
    if "%" in text:
        score += 2
    if "2024" in text:
        score += 2
    if "שיעור" in text:
        score += 2
    if "לוח" in text or "תרשים" in text:
        score += 1
        
    return score


def search_knowledge_base_ranked(query, exact_phrase=None, k=5, fetch_k=30):
    raw_results = vector_db.similarity_search(query, k=fetch_k)
    
    if exact_phrase:
        ranked_results = sorted(
            raw_results,
            key=lambda result: rank_result(result, exact_phrase),
            reverse=True
        )
    else:
        ranked_results = raw_results
    
    for i, result in enumerate(ranked_results[:k], start=1):
        print(f"\n--- תוצאה {i} ---")
        print(f"דוח: {result.metadata['source']}")
        print(f"עמוד: {result.metadata['page']}")
        print(result.page_content[:1000])
```

```python
# ===============================
# שלב 21: בדיקת חיפוש מדורג
# הסבר: בודק האם קטעים עם נתונים ממשיים עולים לראש
# ===============================

search_knowledge_base_ranked(
    query="מה ידוע על צעירים שאינם עובדים ואינם לומדים בישראל וב-OECD?",
    exact_phrase="אינם עובדים ואינם לומדים",
    k=5,
    fetch_k=30
)
```

```python
# ===============================
# שלב 22: סינון טבלאות ותרשימים
# הסבר: מסיר תוצאות שמתחילות ב"לוח" או "תרשים"
# כי הן פחות מתאימות לתשובות טקסטואליות
# ===============================

def is_bad_chunk(text):
    bad_starts = ["לוח", "תרשים", "■"]
    
    return any(text.strip().startswith(b) for b in bad_starts)
```

```python
# ===============================
# שלב 23: חיפוש נקי ומדויק יותר
# הסבר: משלב חיפוש סמנטי + ביטוי מדויק + סינון תוכן לא איכותי
# ===============================

def search_knowledge_base_clean(query, exact_phrase=None, k=5, fetch_k=30):
    raw_results = vector_db.similarity_search(query, k=fetch_k)
    
    # סינון טבלאות/תרשימים
    filtered = [
        r for r in raw_results 
        if not is_bad_chunk(r.page_content)
    ]
    
    # עדיפות לביטוי מדויק
    if exact_phrase:
        exact = [r for r in filtered if exact_phrase in r.page_content]
        others = [r for r in filtered if exact_phrase not in r.page_content]
        results = exact + others
    else:
        results = filtered
    
    for i, result in enumerate(results[:k], start=1):
        print(f"\n--- תוצאה {i} ---")
        print(f"דוח: {result.metadata['source']}")
        print(f"עמוד: {result.metadata['page']}")
        print(result.page_content[:1000])
```

```python
# ===============================
# שלב 25: יצירת תשובה מסכמת ללא מודל
# הסבר: בוחר קטעים איכותיים בלבד ומציג אותם כתשובה מסודרת
# ===============================

def generate_simple_answer(query):
    results = search_knowledge_base_clean(
        query=query,
        exact_phrase="אינם עובדים ואינם לומדים",
        k=5,
        fetch_k=30
    )
```

```python
# ===============================
# עדכון: החזרת תוצאות במקום רק הדפסה
# ===============================

def search_knowledge_base_clean(query, exact_phrase=None, k=5, fetch_k=30):
    raw_results = vector_db.similarity_search(query, k=fetch_k)
    
    filtered = [
        r for r in raw_results 
        if not is_bad_chunk(r.page_content)
    ]
    
    if exact_phrase:
        exact = [r for r in filtered if exact_phrase in r.page_content]
        others = [r for r in filtered if exact_phrase not in r.page_content]
        results = exact + others
    else:
        results = filtered
    
    return results[:k]
```

```python
# ===============================
# יצירת תשובה מסודרת
# ===============================

def generate_simple_answer(query):
    results = search_knowledge_base_clean(
        query=query,
        exact_phrase="אינם עובדים ואינם לומדים",
        k=3,
        fetch_k=30
    )
    
    print("\n### תשובה מבוססת מקורות:\n")
    
    for i, res in enumerate(results, 1):
        print(f"({i}) עמוד {res.metadata['page']}:")
        print(res.page_content[:500])
        print("\n")
```

```python
# ===============================
# שלב 27: סינון טבלאות בלבד
# הסבר: מסיר קטעים שמכילים בעיקר טבלה מספרית
# ===============================

def is_table_chunk(text):
    keywords = ["לוח", "סך", "גברים", "נשים"]
    return sum(1 for k in keywords if k in text) >= 2
```

```python
# ===============================
# שלב 28: בחירת קטעים עם תוכן מילולי
# הסבר: משאיר רק קטעים שיש בהם משפטים ולא רק מספרים
# ===============================

def filter_textual_chunks(results):
    return [
        r for r in results
        if not is_table_chunk(r.page_content)
    ]
```

```python
# ===============================
# שלב 29: יצירת תשובה קריאה למשתמש
# הסבר: מחזיר תשובה קצרה עם נקודות מרכזיות והפניה לעמודים
# ===============================

def generate_readable_answer(query):
    results = search_knowledge_base_clean(
        query=query,
        exact_phrase="אינם עובדים ואינם לומדים",
        k=10,
        fetch_k=30
    )
    
    results = filter_textual_chunks(results)
    
    print("\n### תשובה:\n")
    
    for res in results[:2]:
        print(f"• {res.page_content[:400]}...")
        print(f"(מקור: עמוד {res.metadata['page']})\n")
```

```python
generate_readable_answer(
    "מה ידוע על צעירים שאינם עובדים ואינם לומדים בישראל?"
)
```

```python
# ===============================
# שלב 31: זיהוי סוג השאלה
# הסבר: קובע האם המשתמש מחפש הגדרה, נתונים, השוואה או סיכום
# ===============================

def detect_query_type(query):
    if any(word in query for word in ["מה ההגדרה", "הגדרה", "מה פירוש", "מה זה"]):
        return "definition"
    
    if any(word in query for word in ["נתונים", "שיעור", "אחוז", "כמה", "לוח", "תרשים"]):
        return "data"
    
    if any(word in query for word in ["השוואה", "להשוות", "פערים", "בין"]):
        return "comparison"
    
    return "summary"
```

```python
# ===============================
# שלב 32: התאמת תוצאות לסוג השאלה
# הסבר: לא מסנן הכול, אלא בוחר איזה סוג מידע להעדיף לפי השאלה
# ===============================

def filter_by_query_type(results, query):
    query_type = detect_query_type(query)
    
    if query_type == "definition":
        return results
    
    if query_type == "data":
        return results
    
    if query_type == "comparison":
        return results
    
    if query_type == "summary":
        return [
            r for r in results
            if not is_weak_chunk(r.page_content)
        ]
```
